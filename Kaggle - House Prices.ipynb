# 機器學習預測房價

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.ensemble as se
import sklearn.preprocessing as sp
import sklearn.linear_model as lm
import sklearn.metrics as sm
from sklearn.preprocessing import StandardScaler
from numpy.testing import assert_almost_equal
from sklearn.decomposition import PCA

data_train=pd.read_csv('train.csv')
data_test=pd.read_csv('test.csv')

#取出房價
train_y=data_train['SalePrice']

data_train=data_train.drop(columns=['SalePrice'])

#合併兩個檔案 
data_all=pd.concat([data_train,data_test])

#標頭排序
data_all=data_all.reset_index(drop=True)

#補缺失值    數字na轉成0  字符na轉成 nn
data_all['LotFrontage']=data_all['LotFrontage'].fillna(0)
data_all['MasVnrArea']=data_all['MasVnrArea'].fillna(0)
data_all['GarageYrBlt']=data_all['GarageYrBlt'].fillna(0)
data_all['BsmtFinSF1']=data_all['BsmtFinSF1'].fillna(0)
data_all['BsmtFinSF2']=data_all['BsmtFinSF2'].fillna(0)
data_all['BsmtUnfSF']=data_all['BsmtUnfSF'].fillna(0)
data_all['TotalBsmtSF']=data_all['TotalBsmtSF'].fillna(0)
data_all['BsmtFullBath']=data_all['BsmtFullBath'].fillna(0)
data_all['BsmtHalfBath']=data_all['BsmtHalfBath'].fillna(0)
data_all['GarageCars']=data_all['GarageCars'].fillna(0)
data_all['GarageArea']=data_all['GarageArea'].fillna(0)

B=['SaleType','Functional','KitchenQual','Exterior2nd','Exterior1st','Utilities','MSZoning','Alley','MasVnrType','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical',
'FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']

data_all[B]=data_all[B].fillna('NN')

# 將int轉成float
F2=['Id','MSSubClass','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','BsmtFinSF1',
'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath',
'BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces',
'GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch',
'PoolArea','MiscVal','MoSold','YrSold']

data_all[F2]=data_all[F2].astype('float')

# 整個轉碼
encoders,train_x=[],[]
for row in data_all:
    encoder=sp.LabelEncoder()
    train_x.append(encoder.fit_transform(data_all[row]))
    encoders.append(encoder)

train_x=pd.DataFrame(train_x).T

BX2=[46,4,62,38,37,43,34,20,3,19,
12,67,59,76,44,17,26,66,77,18,
24,33,40,23,32,56,54,1,57,7,
53,60]

train_x2=train_x[:1460]

# 代入調整後的參數    /後面會有找參數的代碼
modelx2=XGBRegressor(learning_rate= 0.07, n_estimators= 400, max_depth= 4, min_child_weight= 2,seed= 0,subsample= 0.7, colsample_bytree= 0.7, gamma= 0.6, reg_alpha= 2, reg_lambda= 3)
modelx2.fit(train_x2, train_y)

pred_y = modelx2.predict(train_x2)
print(sm.r2_score(train_y, pred_y))  

# 雖然之前有找尋重要的特徵(BX2)，但經過測試使用全部資料再配合調整後的參數所得出的值最高，所以上面的代碼並沒有使用 BX2
```

## 預測房價上傳Kaggle

```python
# xgboost 測試
test_x=train_x[1460:]

pred_y2 = modelx2.predict(test_x)

df=pd.DataFrame(pred_y2, columns=['SalePrice'])

# 抓取id
id_test = list(test_x[0])

# id 轉回
df2=encoders[0].inverse_transform(id_test)
df2=pd.DataFrame(df2, columns=["Id"])

# 將小數點去除
df2=df2.astype('int')

# 合併
ans = df2.join(df)
ans=ans.set_index('Id')

# ans.to_csv('all.csv')
print(ans)
```
